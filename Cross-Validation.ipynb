```python
import random
import csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def read_csv(file):
    with open(file,'r') as f:
        reader = csv.reader(f)
        info = []
        for row in reader:
             info.append(row)

        for datum in info:
            for i in range(len(info[0])-1):
                datum[i] = float(datum[i])
    return info
```
